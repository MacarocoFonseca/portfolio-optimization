```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import random
import yfinance as yf
```

```python
stocks = [
    "AAPL",
    "GOOG",
    "CRM",
    "MSFT",
    "SPY",
    "VGT",
    "META",
    "TSLA",
    "ABBV",
    "ASML",
    "NVDA",
    "CCL",
    "BABA",
    "NSRGY",
    "OMV.VI",
    "VOO",
    "VGT",
    "WMT",
    "KO",
    "NKE",
]
start_date = "2013-01-01"
end_date = datetime.today().strftime("%Y-%m-%d")
```

```python
df_assets = pd.DataFrame()

# Store the adjsuted close price of the sotcks in the dataframe
for stock in stocks:
    df_assets[stock] = yf.download(stock, start=start_date, end=end_date)["Adj Close"]
df_assets
```

```python
# Get daily simple returns of assets
daily_simple_returns = df_assets.pct_change()
daily_simple_returns
```

```python
# Calculate the annualized returns with more than 10 years of date

annualized_returns = daily_simple_returns.mean() * 252
annualized_returns
```

```python
# Sort the date
sorted_annualized_returns = annualized_returns.sort_values(ascending=False)
sorted_annualized_returns
```

```python
# Plot the sorted annualized returns
plt.bar(sorted_annualized_returns.index, sorted_annualized_returns)
plt.ylabel("Annualized Returns")
plt.xlabel("Assets")
plt.title("Annualized Returns of Assets")
plt.xticks(rotation=90)
```
